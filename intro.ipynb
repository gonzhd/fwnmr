## My first FID
Lets define our first FID in a few lines of code

```python
aq = 8
td = 8192
t = np.arange(td)*aq/td
w = 100
T1 = 2
I_t = np.sin(w*t)*np.exp(-t/T1)

plt.plot(t, I_t)
```
